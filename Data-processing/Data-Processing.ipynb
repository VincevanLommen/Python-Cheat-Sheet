# Deel 1: Pandas Series

## Wat is pandas?

Pandas is een Python package voor data manipulation en data analysis.

Je gebruikt pandas voor gestructureerde data, bijvoorbeeld tabellen uit CSV-bestanden, spreadsheets of database-achtige data.

Typische operaties:

- filteren
- selecteren
- groeperen
- aggregeren
- ontbrekende waarden behandelen
- visualiseren

## Wat is een Series?

Een `Series` is een eendimensionale array-like datastructuur.

Kenmerken:

- bevat waarden van meestal hetzelfde type
- elk element heeft een index
- kan gemaakt worden uit een list, dictionary of andere Series
- lijkt op een kolom in een spreadsheet

Belangrijk: veel pandas-operaties maken een nieuw object en laten de originele data ongemoeid.

````python
import pandas as pd

scores = pd.Series([85, 78, 92, 88])
print(scores)
print(scores.index)
print(scores.values)
````

## Series met custom index

Je kan labels gebruiken in plaats van numerieke indexen. Dat maakt data leesbaarder.

````python
scores = pd.Series(
    [85, 78, 92, 88],
    index=['Alice', 'Bob', 'Charlie', 'Dina']
)

print(scores)
print(scores['Alice'])
print(scores.loc['Charlie'])
````

## Series uit dictionary

Bij een dictionary worden de keys automatisch de indexlabels.

````python
student_scores = {
    'Alice': 85,
    'Bob': 78,
    'Charlie': 92,
    'Dina': 88,
}

scores = pd.Series(student_scores)
print(scores)
````

## Nieuwe Series met specifieke indexen

Als je een index opgeeft, kiest pandas alleen die labels. Labels die ontbreken krijgen `NaN`.

````python
scores = pd.Series(student_scores, index=['Alice', 'Charlie', 'Eve'])
print(scores)
````

## Data uitlezen uit Series

Gebruik:

- `series['label']` of `series.loc['label']` voor label-based access
- `series.iloc[position]` voor position-based access
- boolean filtering voor voorwaarden

````python
scores = pd.Series(student_scores)

print(scores['Bob'])
print(scores.iloc[0])
print(scores[scores >= 85])
````

## Data aanpassen in Series

Je kan waarden aanpassen via indexlabels of via een conditie.

````python
scores = pd.Series(student_scores)

scores['Bob'] = 80
scores[scores < 85] = 85
scores.index = ['student1', 'student2', 'student3', 'student4']

print(scores)
````

# Deel 2: Pandas DataFrames

## Wat is een DataFrame?

Een `DataFrame` is een tweedimensionale gelabelde datastructuur.

Je kan het zien als:

- een tabel
- een spreadsheet
- een database table

Een DataFrame heeft:

- rows met een index
- columns met kolomnamen
- flexibele selectie op labels of posities

## DataFrame maken uit dictionary

Bij een dictionary worden de keys kolomnamen en de values kolomdata.

````python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Dina'],
    'Age': [21, 25, 30, 28],
    'City': ['Leuven', 'Brussels', 'Leuven', 'Antwerp'],
}

df = pd.DataFrame(data)
print(df)
````

## DataFrame maken uit list of lists

Zonder `columns` gebruikt pandas integerposities als kolomnamen. Met `columns` geef je betekenisvolle namen.

````python
rows = [
    ['Alice', 21, 'Leuven'],
    ['Bob', 25, 'Brussels'],
    ['Charlie', 30, 'Leuven'],
]

df = pd.DataFrame(rows, columns=['Name', 'Age', 'City'])
print(df)
````

## Custom row indexes

Je kan ook rijlabels instellen. Dat maakt `.loc` vaak duidelijker.

````python
df = pd.DataFrame(
    data,
    index=['person1', 'person2', 'person3', 'person4']
)

print(df)
````

## Kolommen en rijen hernoemen

- `df.columns = [...]` vervangt alle kolomnamen.
- `df.rename(columns={...})` wijzigt enkele kolommen.
- `df.rename(index={...})` wijzigt enkele rijlabels.
- Met `inplace=True` wijzig je de bestaande DataFrame, anders krijg je een nieuwe.

````python
df = pd.DataFrame(data, index=['person1', 'person2', 'person3', 'person4'])

renamed = df.rename(columns={'Name': 'FullName'}, index={'person1': 'student1'})

print(df)
print('---')
print(renamed)
````

## Data selecteren

Gebruik:

- `df['Name']` voor een kolom
- `df.loc['person1']` voor een rij via label
- `df.iloc[0]` voor een rij via positie
- `df.at['person1', 'Name']` voor een specifiek element

````python
df = pd.DataFrame(data, index=['person1', 'person2', 'person3', 'person4'])

print(df['Name'])
print('---')
print(df.loc['person1'])
print('---')
print(df.iloc[0])
print('---')
print(df.at['person1', 'Name'])
````

## Slicing en filtering

- `.loc` werkt met labels.
- `.iloc` werkt met integerposities.
- Boolean filtering selecteert rijen die aan een voorwaarde voldoen.

````python
df = pd.DataFrame(data, index=['person1', 'person2', 'person3', 'person4'])

print(df.loc['person2':'person4', 'Name':'Age'])
print('---')
print(df.iloc[1:4, 0:2])
print('---')
print(df[df['Age'] >= 28])
````

## Kolomoperaties: toevoegen, verwijderen en sorteren

Kolommen toevoegen kan met assignment. Verwijderen kan met `drop`. Sorteren kan met `sort_values`.

````python
df = pd.DataFrame(data)

df['Adult'] = df['Age'] >= 18
print(df)

without_city = df.drop(columns=['City'])
print('---')
print(without_city)

sorted_by_age = df.sort_values('Age', ascending=False)
print('---')
print(sorted_by_age)
````

## Missing data

Ontbrekende data wordt vaak weergegeven als `NaN`.

- `dropna()` verwijdert rijen met ontbrekende waarden.
- `fillna(value)` vult ontbrekende waarden op.

Standaard verandert de originele DataFrame niet.

````python
df_missing = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Score': [85, None, 92],
})

print(df_missing)
print('--- dropna')
print(df_missing.dropna())
print('--- fillna')
print(df_missing.fillna(0))
````

## Arithmetic en aggregation

Je kan kolommen optellen, aftrekken, enz. Aggregaties zoals `sum`, `min`, `max` en `mean` werken op kolommen of volledige DataFrames.

````python
grades = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Exam': [85, 78, 92],
    'Project': [90, 80, 88],
})

grades['Total'] = grades['Exam'] + grades['Project']
grades['Average'] = grades[['Exam', 'Project']].mean(axis=1)

print(grades)
print(grades['Exam'].max())
print(grades[['Exam', 'Project']].mean())
````

## Grouping

Met `groupby` groepeer je rijen op basis van een kolom en bereken je aggregaties per groep.

````python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Dina'],
    'City': ['Leuven', 'Brussels', 'Leuven', 'Antwerp'],
    'Score': [85, 78, 92, 88],
})

by_city = df.groupby('City')['Score'].agg(['count', 'mean', 'max'])
print(by_city)
````

## Handige DataFrame attributes/methodes

- `head(x)`: eerste x rijen.
- `info()`: info over kolommen, types en null values.
- `size`: totaal aantal cellen.
- `count()`: aantal niet-null waarden per kolom.

````python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Score': [85, None, 92],
})

print(df.head(2))
print('size:', df.size)
print('count:')
print(df.count())
print('info:')
df.info()
````

## DataFrames mergen

Mergen combineert DataFrames op basis van een gemeenschappelijke kolom.

````python
students = pd.DataFrame({
    'StudentId': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
})

scores = pd.DataFrame({
    'StudentId': [1, 2, 3],
    'Score': [85, 78, 92],
})

merged = pd.merge(students, scores, on='StudentId')
print(merged)
````

## CSV lezen

Met `pd.read_csv` lees je CSV-data rechtstreeks in een DataFrame.

In een echt project geef je een bestandspad mee. In dit voorbeeld gebruiken we `StringIO` om een klein CSV-bestand in de notebook te simuleren.

````python
from io import StringIO

csv_data = StringIO("""Name,Age,City
Alice,21,Leuven
Bob,25,Brussels
Charlie,30,Leuven
""")

df = pd.read_csv(csv_data)
print(df)
````

# Deel 3: Visualisatie met matplotlib

## Waarom visualiseren?

Na data manipulation en analysis wil je vaak patronen zichtbaar maken.

Matplotlib is een visualization library. Pandas integreert goed met matplotlib, waardoor je DataFrames en Series makkelijk kan plotten.

## Figure, Axes, Axis en Artist

- Figure: canvas waarop grafieken staan.
- Axes: gebied waarin data geplot wordt.
- Axis: x-as of y-as met schaal, ticks en labels.
- Artist: alles wat zichtbaar is op de figure, zoals lijnen, tekst en legendes.

````python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])
ax.set_title('Simple line plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.close(fig)
print('figure created')
````

## Meerdere Axes

Een Figure kan meerdere Axes bevatten, bijvoorbeeld een 2x2 grid.

````python
fig, axs = plt.subplots(2, 2, figsize=(6, 4))

for row in axs:
    for ax in row:
        ax.plot([1, 2, 3], [1, 4, 9])

plt.close(fig)
print('2x2 figure created')
````

## Explicit vs implicit interface

- Explicit Axes interface: je werkt met `fig` en `ax` objecten. Dit is duidelijker voor grotere plots.
- Implicit pyplot interface: je gebruikt `plt.plot`, `plt.title`, enz. Dit wordt vaak gebruikt in P2 en simpele voorbeelden.

````python
# Implicit pyplot interface
plt.figure(figsize=(5, 3))
plt.plot([1, 2, 3], [10, 20, 15])
plt.title('Implicit plot')
plt.xlabel('Step')
plt.ylabel('Value')
plt.grid(True)
plt.close()
print('implicit plot created')
````

## Line graph uit DataFrame

Voorbeeld uit de slides: maandelijkse sales en expenses plotten.

````python
import pandas as pd
import matplotlib.pyplot as plt

sales_df = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [100, 120, 150, 130, 140],
    'Expenses': [80, 90, 100, 110, 95],
})

plt.figure(figsize=(10, 6))
plt.plot(
    sales_df['Month'],
    sales_df['Sales'],
    marker='o',
    color='blue',
    linestyle='-',
    linewidth=2,
    markersize=8,
    label='Sales',
)
plt.plot(
    sales_df['Month'],
    sales_df['Expenses'],
    marker='s',
    color='red',
    linestyle='--',
    linewidth=2,
    markersize=8,
    label='Expenses',
)
plt.title('Monthly Sales and Expenses')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.close()
print('sales plot created')
````

## Andere grafieken vanuit DataFrame

Pandas kan zelf plots maken via `.plot(...)`. Achterliggend gebruikt pandas matplotlib.

````python
sales_df = sales_df.set_index('Month')

ax = sales_df.plot(kind='bar', figsize=(6, 4), title='Sales vs Expenses')
ax.set_ylabel('Amount')
plt.close(ax.figure)

ax = sales_df.plot(kind='line', marker='o', figsize=(6, 4), title='Line graph')
plt.close(ax.figure)

print('pandas plots created')
````
