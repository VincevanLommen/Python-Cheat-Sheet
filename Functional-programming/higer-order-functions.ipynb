## 1. Generalizing Functions

Hardcoded waarden uit een functie halen en er een argument van maken.
- minder code-duplicatie
- breder herbruikbaar

````python
# Dubbele logica, enkel de string verschilt
def count_movies_by_spielberg(movies):
    count = 0
    for movie in movies:
        if movie['director'] == 'Steven Spielberg':
            count += 1
    return count

def count_movies_by_coen_brothers(movies):
    count = 0
    for movie in movies:
        if movie['director'] == 'Coen Brothers':
            count += 1
    return count
````

````python
# Gegeneraliseerd: director wordt argument
def count_movies_by_director(movies, director):
    count = 0
    for movie in movies:
        if movie['director'] == director:
            count += 1
    return count

movies = [
    {'title': "Schindler's List", 'director': 'Steven Spielberg'},
    {'title': 'Fargo',            'director': 'Coen Brothers'},
    {'title': 'E.T.',             'director': 'Steven Spielberg'},
]

print(count_movies_by_director(movies, 'Steven Spielberg'))  # 2
print(count_movies_by_director(movies, 'Coen Brothers'))     # 1
````

## 2. Functions as Arguments

In Python zijn functies objecten. Je kan ze opslaan in een variabele of doorgeven als argument.

Zonder haakjes = de functie zelf meegeven. Met haakjes = de functie aanroepen en het resultaat meegeven.

````python
def greet_casually(name):
    print(f"Hi {name}!")

def greet_formally(name):
    print(f"Good day, {name}.")

greet_function = greet_casually   # geen haakjes
greet_function("Mary")            # Hi Mary!

greet_function = greet_formally
greet_function("Mary")            # Good day, Mary.
````

## 3. Higher-Order Functions

Een higher-order function neemt een of meer functies als argument, of geeft een functie terug.

````python
import time, random

def bubble_sort(lst):
    result = list(lst)
    swapped = True
    while swapped:
        swapped = False
        for i in range(len(result) - 1):
            if result[i] > result[i + 1]:
                result[i], result[i + 1] = result[i + 1], result[i]
                swapped = True
    return result

# sorting_function is een functie-argument -> maakt test_sort een HOF
def test_sort(sorting_function):
    for size in [1000, 10000]:
        random_list = random.choices(range(0, 1000), k=size)
        print(f"List size: {size}")
        start = time.time()
        sorting_function(random_list)
        print(f"Time: {time.time() - start:.6f}s\n")

test_sort(bubble_sort)  # bubble_sort meegeven, niet aanroepen
test_sort(sorted)       # Python's ingebouwde sorted
````

## 4. Conditie extraheren

Een if-conditie omzetten naar een functie en als argument meegeven.
Zo krijg je een generieke `count`-functie die met elke conditie werkt.

````python
def is_even(number):
    return number % 2 == 0

def is_child(person):
    return person['age'] < 18

# condition is een functie-argument
def count(collection, condition):
    total = 0
    for element in collection:
        if condition(element):
            total += 1
    return total

print(count([1, 2, 3, 4, 5, 6], is_even))  # 3

people = [
    {'name': 'Alice',   'age': 12},
    {'name': 'Bob',     'age': 25},
    {'name': 'Charlie', 'age': 15},
]
print(count(people, is_child))              # 2
````

## 5. Nested Functions

Wanneer de conditie-functie een extra variabele nodig heeft die niet als argument meegegeven kan worden,
definieer je de hulpfunctie binnenin de buitenste functie.
De nested function heeft dan toegang tot de variabelen van de buitenste scope.

````python
# Werkt niet: director bestaat niet in de scope van is_by_director
def is_by_director(movie):
    return movie['director'] == director  # NameError!
````

````python
# Nested function: is_by_director zit binnenin en ziet 'director'
def count_movies_by_director(movies, director):
    def is_by_director(movie):                # nested function
        return movie['director'] == director  # 'director' uit buitenste scope
    return count(movies, is_by_director)

movies = [
    {'title': 'Once Upon a Time in the West',  'director': 'Sergio Leone'},
    {'title': 'Requiem for a Dream',           'director': 'Darren Aronofsky'},
    {'title': 'The Good, the Bad and the Ugly','director': 'Sergio Leone'},
]

print(count_movies_by_director(movies, 'Sergio Leone'))      # 2
print(count_movies_by_director(movies, 'Darren Aronofsky'))  # 1
````
