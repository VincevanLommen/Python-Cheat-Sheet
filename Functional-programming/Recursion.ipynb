## Wat is recursion?

Recursion is een techniek waarbij een functie zichzelf aanroept.

Een recursive functie heeft altijd twee delen nodig:

- Base case: wanneer stoppen we?
- Recursive case: hoe maken we het probleem kleiner?

Zonder base case blijft de functie zichzelf oproepen tot Python stopt met een recursion error.

````python
def sum_to(n):
    if n == 0:          # base case
        return 0
    return n + sum_to(n - 1)  # recursive case


print(sum_to(5))
````

## Iteration vs recursion

Iteration gebruikt loops. Recursion gebruikt functieaanroepen.

Beide kunnen vaak hetzelfde probleem oplossen. Iteration is meestal beter voor simpele herhaling. Recursion is duidelijker bij problemen die natuurlijk uit kleinere versies van zichzelf bestaan.

````python
def print_numbers_iterative(start, end):
    for number in range(start, end + 1):
        print(number)


def print_numbers_recursive(start, end):
    if start > end:
        return
    print(start)
    print_numbers_recursive(start + 1, end)


print_numbers_iterative(1, 3)
print('---')
print_numbers_recursive(1, 3)
````

## Factorial

`n!` betekent: `n * (n-1) * ... * 1`.

Voorbeeld:

```text
5! = 5 * 4 * 3 * 2 * 1 = 120
```

Base case: `1! = 1`.

````python
def factorial_recursive(n):
    if n < 0:
        raise ValueError('n must be positive')
    if n in (0, 1):
        return 1
    return n * factorial_recursive(n - 1)


print(factorial_recursive(5))
````

## Hoe bouw je een recursive oplossing?

1. Zoek het simpelste geval.
2. Schrijf daarvoor de base case.
3. Bepaal hoe je het probleem kleiner maakt.
4. Combineer het huidige stuk met de oplossing van het kleinere probleem.

````python
def contains_value(values, target):
    if len(values) == 0:
        return False
    if values[0] == target:
        return True
    return contains_value(values[1:], target)


print(contains_value([10, 20, 30], 20))
print(contains_value([10, 20, 30], 99))
````

## Wanneer recursion gebruiken?

Gebruik recursion vooral wanneer het probleem een recursive structuur heeft:

- bomen
- grafen
- linked lists
- divide-and-conquer algoritmes zoals merge sort, quick sort en binary search

Voor simpele sommen of tellers is een loop meestal beter.

````python
class TreeNode:
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right


def contains(node, target):
    if node is None:
        return False
    if node.value == target:
        return True
    return contains(node.left, target) or contains(node.right, target)


root = TreeNode(
    'D',
    left=TreeNode('B', TreeNode('A'), TreeNode('C')),
    right=TreeNode('F', TreeNode('E'), TreeNode('H')),
)

print(contains(root, 'E'))
print(contains(root, 'Z'))
````

## Recursion risico's

- Elke recursive call gebruikt geheugen op de call stack.
- Te diepe recursion geeft `RecursionError`.
- Niet elk probleem wordt duidelijker met recursion.

Gebruik recursion wanneer de structuur van het probleem er echt bij past.
