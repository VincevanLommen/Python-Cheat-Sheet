## Iterables

Een iterable is iets waar je over kan itereren met een `for`-loop.

Voorbeelden:

- list
- tuple
- set
- dictionary
- string
- range

Ze verschillen onderling, maar ze delen een eigenschap: Python kan er element per element door gaan.

````python
collections = [
    [1, 2, 3],
    (1, 2, 3),
    {'a', 'b', 'c'},
    {'name': 'Alice', 'score': 85},
    'abc',
]

for collection in collections:
    print(type(collection).__name__, list(collection))
````

## Generator functions

Een generator function gebruikt `yield` in plaats van `return`.

Waarom?

- Een gewone functie geeft een resultaat terug en stopt.
- Een generator kan meerdere waarden produceren.
- De functie pauzeert bij `yield` en gaat pas verder wanneer de volgende waarde nodig is.

Hoe?

```python
def generator():
    yield waarde
```

````python
def countdown(start):
    current = start
    while current > 0:
        yield current
        current -= 1


for number in countdown(3):
    print(number)
````

## Lazy evaluation

Een generator voert niet alles meteen uit. De body draait alleen wanneer je een volgende waarde vraagt.

Dit is nuttig wanneer:

- je maar een deel van de waarden nodig hebt
- de reeks heel groot is
- de reeks oneindig kan zijn

````python
def noisy_generator():
    print('start')
    yield 1
    print('middle')
    yield 2
    print('end')


values = noisy_generator()
print('Generator gemaakt')
print(next(values))
print(next(values))
````

## `next()`

Met `next(generator)` vraag je handmatig de volgende waarde op. Als er geen waarde meer is, krijg je `StopIteration`.

````python
values = countdown(2)

print(next(values))
print(next(values))

try:
    print(next(values))
except StopIteration:
    print('Geen waarden meer')
````

## Generators vs lists

Een list bewaart alle waarden in geheugen. Een generator produceert waarden een voor een.

Waarom?

Voor grote reeksen kan een generator veel minder geheugen gebruiken.

````python
import sys

numbers_list = [number for number in range(100_000)]
numbers_generator = (number for number in range(100_000))

print(sys.getsizeof(numbers_list))
print(sys.getsizeof(numbers_generator))
````

## Generators kan je maar een keer doorlopen

Als een generator opgebruikt is, zijn de waarden weg. Wil je opnieuw starten, dan maak je een nieuwe generator.

````python
values = countdown(3)

print(list(values))
print(list(values))

print(list(countdown(3)))
print(list(countdown(3)))
````

## Infinite generators

Een generator kan oneindig veel waarden produceren, omdat hij niet alles in geheugen moet opslaan.

Let op: combineer dit altijd met een stopconditie in je code.

````python
def natural_numbers():
    number = 1
    while True:
        yield number
        number += 1


numbers = natural_numbers()
for _ in range(5):
    print(next(numbers))
````

## Generator comprehensions

Een generator comprehension lijkt op een list comprehension, maar gebruikt ronde haakjes.

```python
(number ** 2 for number in numbers)
```

Waarom?

- Je hebt de volledige lijst niet nodig.
- Je geeft de waarden meteen door aan functies zoals `sum`, `min`, `max`, `all` of `any`.

````python
numbers = range(1, 6)

squares_list = [number ** 2 for number in numbers]
squares_generator = (number ** 2 for number in numbers)

print(squares_list)
print(list(squares_generator))
````

````python
numbers = range(1, 1_000_001)

print(sum(number for number in numbers if number % 2 == 0))
print(any(number > 999_990 for number in range(1_000_000)))
print(all(number >= 0 for number in range(10)))
````

## `itertools`

`itertools` bevat hulpmiddelen voor werken met iterables en generators. Handig om combinaties, ketens of stukken van iterables te maken.

````python
import itertools

letters = ['A', 'B']
numbers = [1, 2, 3]

print(list(itertools.product(letters, numbers)))
print(list(itertools.islice(natural_numbers(), 5, 10)))
````
