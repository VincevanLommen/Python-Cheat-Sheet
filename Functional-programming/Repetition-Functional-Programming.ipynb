Herhalingsoefening waarin functional programming en recursion samenkomen.

## Gebruikte concepten

- List comprehensions
- Higher-order functions
- Lambda functions
- `sorted` met `key`
- Recursion op een binary tree

## Dataset

We gebruiken een lijst van dictionaries. Elke dictionary stelt een studentresultaat voor.

````python
students = [
    {'name': 'Alice', 'subject': 'Mathematics', 'score': 85},
    {'name': 'Bob', 'subject': 'Physics', 'score': 78},
    {'name': 'Charlie', 'subject': 'Mathematics', 'score': 92},
    {'name': 'Dina', 'subject': 'Chemistry', 'score': 88},
    {'name': 'Eve', 'subject': 'Mathematics', 'score': 76},
]

print(students)
````

## Scores verzamelen zonder comprehension

Eerst de klassieke manier met een lege lijst en `append`.

````python
def collect_scores_without_comprehension(students):
    scores = []
    for student in students:
        scores.append(student['score'])
    return scores


print(collect_scores_without_comprehension(students))
````

## Scores verzamelen met list comprehension

Dezelfde logica, maar korter en duidelijker voor dit simpele mapping-patroon.

````python
def collect_scores(students):
    return [student['score'] for student in students]


print(collect_scores(students))
````

## Studenten filteren zonder higher-order function

Dit werkt, maar de conditie is hardcoded in de functie.

````python
def filter_high_scores(students):
    result = []
    for student in students:
        if student['score'] >= 80:
            result.append(student)
    return result


print(filter_high_scores(students))
````

## Filteren met higher-order function

We maken de filterfunctie herbruikbaar door de conditie als functieargument mee te geven.

````python
def filter_students(students, condition):
    return [student for student in students if condition(student)]


high_scores = filter_students(students, lambda student: student['score'] >= 80)
math_students = filter_students(students, lambda student: student['subject'] == 'Mathematics')

print(high_scores)
print(math_students)
````

## Filteren en sorteren combineren

Opdracht uit de slides: neem alleen studenten van `Mathematics` en sorteer ze op score van hoog naar laag.

````python
math_students = filter_students(students, lambda student: student['subject'] == 'Mathematics')
math_students_by_score = sorted(math_students, key=lambda student: student['score'], reverse=True)

print(math_students_by_score)
````

## Extra: generieke sorteerfunctie

Ook sorteren kan je generaliseren door de `key` mee te geven.

````python
def sort_students(students, key_function, reverse=False):
    return sorted(students, key=key_function, reverse=reverse)


print(sort_students(students, lambda student: student['name']))
print(sort_students(students, lambda student: student['score'], reverse=True))
````

## Recursion herhaling: binary tree

We bouwen een boom met root `D`:

```text
        D
      /        B     F
    / \   / \
   A   C E   H
```

De functie `contains(node, target)` zoekt recursief in de boom.

````python
class TreeNode:
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right


class BinaryTree:
    def __init__(self, root=None):
        self.root = root

    def contains(self, target):
        return self.__contains(self.root, target)

    def __contains(self, node, target):
        if node is None:
            return False
        if node.value == target:
            return True
        return self.__contains(node.left, target) or self.__contains(node.right, target)


root = TreeNode(
    'D',
    left=TreeNode('B', TreeNode('A'), TreeNode('C')),
    right=TreeNode('F', TreeNode('E'), TreeNode('H')),
)

tree = BinaryTree(root)
print(tree.contains('A'))
print(tree.contains('H'))
print(tree.contains('Z'))
````

## Overzicht

| Probleem | Techniek |
|---|---|
| Waarden uit dictionaries halen | List comprehension |
| Filterlogica herbruikbaar maken | Higher-order function |
| Kleine conditie meegeven | Lambda |
| Sorteren op specifiek veld | `sorted(..., key=...)` |
| Zoeken in boomstructuur | Recursion |
