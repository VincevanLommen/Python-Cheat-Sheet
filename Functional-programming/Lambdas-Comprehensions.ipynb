## Lambda functions

Een lambda is een korte manier om een eenvoudige functie te schrijven.

Waarom?

- Bij higher-order functions heb je vaak kleine helperfuncties nodig.
- Voor simpele logica is een volledige `def` soms te veel code.
- Een lambda kan direct meegegeven worden als argument.

Hoe?

```python
lambda argumenten: expressie
```

Belangrijk: een lambda bevat maar een expressie. Voor meerdere stappen gebruik je een gewone functie.

````python
double = lambda number: number * 2
is_even = lambda number: number % 2 == 0

print(double(5))
print(is_even(6))
print(is_even(7))
````

## Lambda als argument

Omdat een lambda een expressie is, kan je ze direct meegeven aan een functie.

````python
def count(collection, condition):
    total = 0
    for item in collection:
        if condition(item):
            total += 1
    return total


numbers = [1, 2, 3, 4, 5, 6]
print(count(numbers, lambda number: number % 2 == 0))
print(count(numbers, lambda number: number > 3))
````

## Lambda vs nested function

Een lambda gedraagt vaak zoals een nested function: ze wordt gemaakt op de plaats waar je ze nodig hebt en kan variabelen uit die scope gebruiken.

Gebruik een lambda alleen als de logica kort en duidelijk blijft.

````python
movies = [
    {'title': 'Fargo', 'director': 'Coen Brothers'},
    {'title': 'E.T.', 'director': 'Steven Spielberg'},
    {'title': 'Jaws', 'director': 'Steven Spielberg'},
]


def count_movies_by_director(movies, director):
    return count(movies, lambda movie: movie['director'] == director)


print(count_movies_by_director(movies, 'Steven Spielberg'))
````

## Sorteren met `key`

`sorted()` en `.sort()` gebruiken standaard `<` om te vergelijken. Met `key` geef je een functie mee die bepaalt waarop gesorteerd wordt.

Waarom?

- Strings sorteren standaard alfabetisch, maar soms wil je sorteren op lengte.
- Dictionaries/objecten hebben niet automatisch een duidelijke sorteerwaarde.
- Je kan meerdere sorteercriteria combineren met tuples.

````python
names = ['Ada', 'Guido', 'Linus', 'Grace']

print(sorted(names))
print(sorted(names, key=lambda name: len(name)))
````

````python
students = [
    {'name': 'Alice', 'subject': 'Mathematics', 'score': 85},
    {'name': 'Bob', 'subject': 'Physics', 'score': 78},
    {'name': 'Charlie', 'subject': 'Mathematics', 'score': 92},
]

by_score_desc = sorted(students, key=lambda student: student['score'], reverse=True)
print(by_score_desc)

by_subject_then_score = sorted(students, key=lambda student: (student['subject'], -student['score']))
print(by_subject_then_score)
````

## `min` en `max` met `key`

`min()` en `max()` werken op dezelfde manier als sorteren: standaard vergelijken ze de waarden zelf, maar met `key` kies je waarop vergeleken wordt.

````python
students = [
    {'name': 'Alice', 'score': 85},
    {'name': 'Bob', 'score': 78},
    {'name': 'Charlie', 'score': 92},
]

best_student = max(students, key=lambda student: student['score'])
lowest_student = min(students, key=lambda student: student['score'])

print(best_student)
print(lowest_student)
````

## List comprehensions

Een list comprehension is een korte manier om een nieuwe lijst te maken op basis van een bestaande iterable.

Waarom?

- Minder boilerplate dan `for` + `append`.
- Duidelijk voor simpele map/filter patronen.
- Vaak efficienter en idiomatischer in Python.

Vorm:

```python
[expressie for item in collectie]
[expressie for item in collectie if conditie]
```

````python
numbers = [1, 2, 3, 4, 5]

squares = [number ** 2 for number in numbers]
even_numbers = [number for number in numbers if number % 2 == 0]
even_squares = [number ** 2 for number in numbers if number % 2 == 0]

print(squares)
print(even_numbers)
print(even_squares)
````

## Mapping en filtering combineren

Links van `for` staat wat je in de nieuwe lijst stopt. Rechts met `if` bepaal je welke elementen mogen meedoen.

````python
students = [
    {'name': 'Alice', 'subject': 'Mathematics', 'score': 85},
    {'name': 'Bob', 'subject': 'Physics', 'score': 78},
    {'name': 'Charlie', 'subject': 'Mathematics', 'score': 92},
    {'name': 'Dina', 'subject': 'Chemistry', 'score': 88},
]

math_names = [student['name'] for student in students if student['subject'] == 'Mathematics']
high_score_names = [student['name'] for student in students if student['score'] >= 80]

print(math_names)
print(high_score_names)
````

## Set comprehensions

Een set comprehension gebruikt `{}` en verwijdert automatisch duplicaten. De volgorde is niet gegarandeerd.

````python
subjects = ['Math', 'Physics', 'Math', 'Chemistry', 'Physics']
unique_subjects = {subject for subject in subjects}

print(unique_subjects)
````

## Dictionary comprehensions

Een dictionary comprehension gebruikt ook `{}`, maar je schrijft `key: value`.

````python
students = [
    {'name': 'Alice', 'score': 85},
    {'name': 'Bob', 'score': 78},
    {'name': 'Charlie', 'score': 92},
]

scores_by_name = {student['name']: student['score'] for student in students}
print(scores_by_name)
````

## Nested comprehensions en flattening

Bij nested loops schrijf je de `for`-delen in dezelfde volgorde als gewone loops.

Flattening betekent: een lijst van lijsten omzetten naar een platte lijst.

````python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9],
]

flattened = [number for row in matrix for number in row]
print(flattened)

coordinates = [(x, y) for x in range(2) for y in range(3)]
print(coordinates)
````

## Handige built-ins

- `all(iterable)`: `True` als alle waarden truthy zijn. Voor een lege iterable is dit `True`.
- `any(iterable)`: `True` als minstens een waarde truthy is. Voor een lege iterable is dit `False`.
- `zip(a, b)`: koppelt elementen op dezelfde positie.
- `enumerate(iterable)`: koppelt elk element aan een index.

````python
scores = [85, 78, 92]

print(all(score >= 50 for score in scores))
print(any(score >= 90 for score in scores))

names = ['Alice', 'Bob', 'Charlie']
for name, score in zip(names, scores):
    print(name, score)

for index, name in enumerate(names, start=1):
    print(index, name)
````
