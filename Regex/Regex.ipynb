# Regex: regular expressions in Python

Complete cheat sheet voor de twee regexdelen.

## Wat is regex?

Regex staat voor regular expressions: een compacte taal om patronen in strings te beschrijven.

Waarom?

- Zoeken naar tekstpatronen.
- Informatie uit strings halen.
- Tekst vervangen op basis van een patroon.
- Input valideren, bijvoorbeeld email, telefoonnummer of ticketcode.

Hoe?

In Python gebruik je de ingebouwde module `re`. Gebruik meestal raw strings voor patronen: `r'\d+'`. Dan behandelt Python backslashes niet eerst als gewone escape characters.

```python
import re

text = 'The cat and the hat'
pattern = r'cat'

match = re.search(pattern, text)
if match:
    print(match.group())
```

## Belangrijkste `re` functies

| Functie | Betekenis |
|---|---|
| `re.match(pattern, string)` | matcht alleen aan het begin van de string |
| `re.fullmatch(pattern, string)` | de volledige string moet matchen |
| `re.search(pattern, string)` | zoekt de eerste match ergens in de string |
| `re.findall(pattern, string)` | geeft alle niet-overlappende matches terug |
| `re.sub(pattern, repl, string)` | vervangt matches door nieuwe tekst |

```python
import re

text = 'cat and dog and cat'

print(re.match(r'cat', text).group())
print(re.search(r'dog', text).group())
print(re.findall(r'cat', text))
print(re.sub(r'cat', 'hamster', text))
print(re.fullmatch(r'cat', 'cat').group())
print(re.fullmatch(r'cat', 'cats'))
```

## `.` any character

`.` matcht eender welk karakter, behalve een newline.

```python
import re

text = 'abc123'
match = re.search(r'a.c', text)
print(match.group())
```

## `^` start van string

`^` matcht het begin van de string. Handig wanneer een patroon verplicht vooraan moet staan.

```python
import re

print(bool(re.search(r'^start', 'start with this')))
print(bool(re.search(r'^start', 'do not start with this')))
```

## `$` einde van string

`$` matcht het einde van de string.

```python
import re

print(bool(re.search(r'this$', 'end with this')))
print(bool(re.search(r'this$', 'this is not the end')))
```

## `*`, `+` en `?`

Deze quantifiers bepalen hoeveel keer het vorige stuk mag voorkomen.

| Pattern | Betekenis |
|---|---|
| `*` | nul of meer keer |
| `+` | een of meer keer |
| `?` | nul of een keer |

```python
import re

for text in ['ac', 'abc', 'abbbbbc']:
    print(text, bool(re.fullmatch(r'ab*c', text)))

print(bool(re.fullmatch(r'ab+c', 'ac')))
print(bool(re.fullmatch(r'ab+c', 'abc')))
print(bool(re.fullmatch(r'ab?c', 'ac')))
print(bool(re.fullmatch(r'ab?c', 'abc')))
print(bool(re.fullmatch(r'ab?c', 'abbc')))
```

## `{m}`, `{m,n}` exact of bereik

Met curly braces geef je een exact aantal of bereik aan.

- `{3}` exact 3 keer
- `{2,4}` minimaal 2 en maximaal 4 keer
- `{2,}` minimaal 2 keer

```python
import re

print(bool(re.fullmatch(r'a{3}', 'aaa')))
print(bool(re.fullmatch(r'a{3}', 'aa')))
print(bool(re.fullmatch(r'a{2,4}', 'aaa')))
print(bool(re.fullmatch(r'a{2,4}', 'aaaaa')))

text = 'aaabbbccc'
match = re.search(r'a{2,3}b{2,3}c{2,3}', text)
print(match.group())
```

## `[]` character class

Square brackets matchen een karakter uit de set.

Voorbeelden:

- `[aeiou]` matcht een klinker.
- `[a-z]` matcht een kleine letter.
- `[0-9]` matcht een cijfer.
- `[^0-9]` matcht alles behalve een cijfer.

```python
import re

print(re.search(r'[aeiou]', 'apple').group())
print(re.findall(r'[0-9]', 'a1b2c3'))
print(re.findall(r'[^0-9]', 'a1b2'))
```

## `()` groups

Parentheses groeperen een deel van je patroon. Met `group(1)`, `group(2)`, ... kan je submatches opvragen.

```python
import re

text = 'abc123'
match = re.search(r'(abc)(123)', text)

print(match.group(0))
print(match.group(1))
print(match.group(2))
print(match.groups())
```

## `|` OR

`|` betekent: match links of rechts. Gebruik haakjes om duidelijk te maken wat bij de OR hoort.

```python
import re

text = 'cat or dog'
print(re.search(r'cat|dog', text).group())

print(bool(re.fullmatch(r'(cat|dog)', 'cat')))
print(bool(re.fullmatch(r'(cat|dog)', 'dog')))
print(bool(re.fullmatch(r'(cat|dog)', 'bird')))
```

## `match` vs `search` vs `fullmatch`

- `match`: alleen vanaf het begin.
- `search`: ergens in de string.
- `fullmatch`: de volledige string moet voldoen.

Voor validatie gebruik je vaak `fullmatch`, omdat je niet alleen een stukje van de input wil goedkeuren.

```python
import re

pattern = r'cat'

print(re.match(pattern, 'cat and dog').group())
print(re.match(pattern, 'dog and cat'))
print(re.search(pattern, 'dog and cat').group())
print(re.fullmatch(pattern, 'cat'))
print(re.fullmatch(pattern, 'cats'))
```

## `findall`

`findall` geeft alle niet-overlappende matches terug.

```python
import re

text = 'apple banana cherry pear'
print(re.findall(r'\b\w{5}\b', text))
print(re.findall(r'\b\w{6}\b', text))
```

## `sub`

`sub` vervangt matches door een replacement string. Alleen niet-overlappende matches worden vervangen.

```python
import re

text = 'Hello, my name is John. John is 20 years old.'
print(re.sub(r'John', 'Alice', text))

print(re.sub(r'aa', 'b', 'aaa'))
```

# Deel 2: Character classes, groups en backreferences

## Raw strings

Gebruik voor regex meestal raw strings: `r'pattern'`.

Waarom?

Regex gebruikt veel backslashes, zoals `\d`, `\w` en `\b`. Met een raw string blijft de backslash zichtbaar voor de regex engine.

```python
import re

path = r'C:\Users\harry'
print(path)

print(bool(re.fullmatch(r'\d{3}', '123')))
print(bool(re.fullmatch(r'\d{3}', 'abc')))
```

## Predefined character classes

| Pattern | Betekenis |
|---|---|
| `\d` | digit, hetzelfde als `[0-9]` |
| `\D` | geen digit, hetzelfde als `[^0-9]` |
| `\w` | word character: letters, cijfers en `_` |
| `\W` | geen word character |
| `\b` | word boundary, begin/einde van woord |
| `\B` | geen word boundary |
| `\\` | letterlijke backslash |

```python
import re

text = 'abc 123 _ !'

print(re.findall(r'\d', text))
print(re.findall(r'\D', text))
print(re.findall(r'\w', text))
print(re.findall(r'\W', text))
```

## Word boundaries `\b`

`\b` matcht een grens tussen woord-karakter en niet-woord-karakter.

Handig om volledige woorden te vinden in plaats van stukjes binnen woorden.

```python
import re

text = 'cat category bobcat cat'

print(re.findall(r'cat', text))
print(re.findall(r'\bcat\b', text))
```

## Email voorbeeld stap voor stap

We bouwen geen perfecte emailvalidator, maar een nuttig patroon voor de oefeningen.

Voorbeeldvragen:

- Start de email met een letter?
- Eindigt de email op `.com` of `.be`?
- Bevat de email een `@` met tekst ervoor en erna?

```python
import re

email = 'harry.potter@hogwarts.com'

print(bool(re.match(r'[a-zA-Z]', email)))
print(bool(re.fullmatch(r'[a-zA-Z].*', email)))
print(bool(re.fullmatch(r'.*\.com$', email)))
print(bool(re.fullmatch(r'.*(\.com|\.be)$', email)))
print(bool(re.fullmatch(r'.+@.+', email)))
```

## Telefoonnummer voorbeeld

`\d{10,13}` betekent 10 tot 13 cijfers. Gebruik `fullmatch` als de hele string een telefoonnummer moet zijn.

```python
import re

phone = '0032478123456'

print(bool(re.fullmatch(r'\d{10,13}', phone)))
print(bool(re.fullmatch(r'\d{10}|\d{13}', phone)))
print(bool(re.fullmatch(r'\d{10}|\d{13}', '0478123456')))
print(bool(re.fullmatch(r'\d{10}|\d{13}', '12345678901')))
```

## Negation met `^` binnen `[]`

Buiten `[]` betekent `^` start van string. Binnen `[]` betekent `^` negatie.

Voorbeeld: `[^0-9]` matcht een karakter dat geen cijfer is.

```python
import re

text = 'a1b2!'
print(re.findall(r'[^0-9]', text))
print(re.findall(r'^[a-z]', text))
```

## Subgroups uitlezen

Alles tussen `()` is een group.

- `group(0)` is de volledige match.
- `group(1)` is de eerste group.
- `groups()` geeft alle subgroups als tuple.

```python
import re

email = 'harry.potter@hogwarts.com'
pattern = r'(.+)\.(.+)@.+'
match = re.fullmatch(pattern, email)

print(match.groups())
print(match.group(0))
print(match.group(1))
print(match.group(2))
```

## Greedy matching

Quantifiers zoals `.+` zijn greedy: ze nemen zoveel mogelijk tekst.

Gebruik specifiekere character classes als je exactere groups wil.

```python
import re

email = 'harry.potter@hogwarts.com'

match_greedy = re.fullmatch(r'(.+)\.(.+)@(.+)', email)
match_specific = re.fullmatch(r'([a-z]+)\.([a-z]+)@([a-z]+)\.([a-z]+)', email)

print(match_greedy.groups())
print(match_specific.groups())
```

## Backreferences

Een backreference verwijst naar tekst die eerder door een group gematcht werd.

- `(\b\w+\b)` matcht een woord en bewaart het als group 1.
- `\1` matcht exact dezelfde tekst als group 1.

Zo kan je dubbele woorden vinden.

```python
import re

text = 'We confirmed that that Harry needs to to call Ron.'
pattern = r'(\b\w+\b) \1'

print(re.findall(pattern, text))
print(re.sub(pattern, r'\1', text))
```

## Emails zoeken in langere tekst

`search` geeft alleen de eerste match. `findall` geeft alle matches.

```python
import re

text = (
    'Call +32478123456 or email harry.potter@hogwarts.com. '
    'If he does not respond, email ron.weasley@hogwarts.com.'
)

pattern = r'[\w.]+@[\w]+\.[\w]+'

print(re.search(pattern, text).group())
print(re.findall(pattern, text))
```

## Alle `search` resultaten met een while-loop

Dit is minder kort dan `findall`, maar nuttig als je ook `span()` of match-objecten nodig hebt.

```python
import re

text = (
    'Email harry.potter@hogwarts.com first. '
    'Then try ron.weasley@hogwarts.com.'
)
pattern = r'[\w.]+@[\w]+\.[\w]+'

remaining = text
match = re.search(pattern, remaining)
while match is not None:
    print(match.group(), match.span())
    remaining = remaining[match.end():]
    match = re.search(pattern, remaining)
```

## Praktische validatiefuncties

In oefeningen is het handig om regex in kleine functies te steken. Dan kan je die functies makkelijk testen.

```python
import re


def is_email(value):
    return re.fullmatch(r'[\w.]+@[\w]+\.[\w]+', value) is not None


def is_belgian_mobile(value):
    return re.fullmatch(r'(\+32|0032)\d{9}', value) is not None


print(is_email('harry.potter@hogwarts.com'))
print(is_email('not an email'))
print(is_belgian_mobile('+32478123456'))
print(is_belgian_mobile('0478123456'))
```
