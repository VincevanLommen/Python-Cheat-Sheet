## Abstract class

Een abstract class is een klasse die bedoeld is als blueprint. Je maakt er normaal geen objecten van.

Waarom?

- Je wil een gemeenschappelijk contract afdwingen.
- Subclasses moeten bepaalde methodes of properties implementeren.
- Je voorkomt dat een onvolledig basisconcept als object gebruikt wordt.

Hoe?

Gebruik `ABC` en `@abstractmethod` uit de module `abc`.

```python
from abc import ABC, abstractmethod


class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass


class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def calculate_area(self):
        return self.width * self.height


rectangle = Rectangle(5, 10)
print(rectangle.calculate_area())
```

## Wat gebeurt er als een subclass de abstract method niet implementeert?

Python laat je dan geen object maken van die subclass. Dat is nuttig, want het probleem wordt snel zichtbaar.

```python
from abc import ABC, abstractmethod


class PaymentMethod(ABC):
    @abstractmethod
    def pay(self, amount):
        pass


class Bancontact(PaymentMethod):
    pass


try:
    method = Bancontact()
except TypeError as error:
    print(error)
```

## Polymorphism met abstract methods

Polymorphism betekent dat verschillende objecten dezelfde methode hebben, maar elk hun eigen implementatie gebruiken.

Waarom?

Je code kan werken met het algemene type, bijvoorbeeld `Shape`, zonder te weten of het echt een `Circle` of `Rectangle` is.

```python
from abc import ABC, abstractmethod


class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return 3.14 * self.radius ** 2


class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def calculate_area(self):
        return self.width * self.height


shapes = [Circle(7), Rectangle(5, 10)]

for shape in shapes:
    print(shape.calculate_area())
```

## Abstract properties

Een abstract property verplicht subclasses om een property met een bepaalde naam te voorzien.

Belangrijk: bij een abstract getter staat `@property` boven `@abstractmethod`.

Waarom?

- Je wil dat elke subclass dezelfde property heeft.
- Elke subclass mag zelf bepalen waar de waarde vandaan komt.
- Je kan validatie afdwingen in de concrete setter.

```python
from abc import ABC, abstractmethod


class Shape(ABC):
    @property
    @abstractmethod
    def color(self):
        pass

    @color.setter
    @abstractmethod
    def color(self, value):
        pass

    @abstractmethod
    def calculate_area(self):
        pass


class ColoredRectangle(Shape):
    def __init__(self, width, height, color):
        self.width = width
        self.height = height
        self.color = color

    @property
    def color(self):
        return self.__color

    @color.setter
    def color(self, value):
        if not isinstance(value, str) or len(value) == 0:
            raise ValueError("Color must be a non-empty string")
        self.__color = value

    def calculate_area(self):
        return self.width * self.height


rectangle = ColoredRectangle(5, 10, "blue")
print(rectangle.color)
print(rectangle.calculate_area())
```

## Abstract property met read-only gedrag

Als je in de abstract class alleen een getter definieert, verplicht je subclasses om minstens leesbare property te voorzien.

```python
from abc import ABC, abstractmethod


class User(ABC):
    @property
    @abstractmethod
    def display_name(self):
        pass


class Student(User):
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name

    @property
    def display_name(self):
        return f"{self.first_name} {self.last_name}"


student = Student("Vince", "van Lommen")
print(student.display_name)
```
