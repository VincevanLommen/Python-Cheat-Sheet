# Properties in Python

Korte uitleg + voorbeelden voor je cheat sheet.

## Wat zijn properties?
Properties laten je toe om **attributen te gebruiken alsof ze publiek zijn**, maar ondertussen **logica, validatie of berekeningen** toe te voegen.

Er zijn drie hoofdtypes:
1. Read‑only properties
2. Computed properties
3. Data validation via setters


# Properties: setter

Voorbeeld met validatie op temperatuur.

```python
class Tea:
    def __init__(self, name, temperature):
        self.name = name
        self.temperature = temperature  # setter wordt gebruikt

    @property
    def temperature(self):
        return self.__temperature

    @temperature.setter
    def temperature(self, value):
        if value <= 0 or value >= 100:
            raise ValueError("Tea must be liquid")
        self.__temperature = value

green = Tea("Green", 80)
print(green.name)
print(green.temperature)

green.name = "Green Tea"
green.temperature = 82
```

# Properties: remarks

- `name` blijft publiek
- `temperature` blijft privé
- Beide worden gebruikt alsof het gewone attributen zijn
- De echte waarde zit in `Tea.__temperature`

```python
green = Tea("Green", 80)
print(green.name)
print(green.temperature)

green.name = "Green Tea"
green.temperature = 82
```

# Read‑only properties

Een property zonder setter kan enkel gelezen worden.

```python
class Vegetable:
    def __init__(self, name, color, taste):
        self.__name = name
        self.color = color
        self.taste = taste

    @property
    def name(self):
        print("reading value")
        return self.__name

veg = Vegetable("Zesty Zucchini", "green", "zesty")
print(veg.name)
```

# Computed properties

Een property die **berekend wordt** op het moment dat je hem opvraagt.

```python
class Vegetable:
    def __init__(self, name, weight):
        self.__name = name
        self.weight = weight

    @property
    def weight_in_pounds(self):
        return self.weight * 2.2

veg = Vegetable("Zesty Zucchini", 0.25)
print(veg.weight_in_pounds)
```

# Data validation met setters

Hier kan je extra logica toevoegen wanneer een waarde wordt aangepast.

```python
class Vegetable:
    def __init__(self, name, color, taste):
        self.name = name
        self.color = color
        self.taste = taste

    @property
    def name(self):
        print("reading value")
        return self.__name

    @name.setter
    def name(self, value):
        print("setting value")
        if value is None or len(value) == 0:
            raise ValueError("No valid name")
        self.__name = value

# Example 1
veg = Vegetable("Zesty Zucchini", "green", "zesty")
veg.name = ""  # zal error geven

# Example 2
another = Vegetable(None, "green", "zesty")  # zal ook error geven
```
