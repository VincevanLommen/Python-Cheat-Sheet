## Overriding

Overriding betekent dat een subclass een methode definieert met dezelfde naam als een methode in de parent class.

Waarom?

- De parent class geeft standaardgedrag.
- De child class kan dat gedrag vervangen of specifieker maken.
- Dit werkt alleen logisch binnen inheritance.

```python
class Animal:
    def speak(self):
        print("Animal speaks")


class Dog(Animal):
    def speak(self):
        print("Dog barks")


animal = Animal()
dog = Dog()

animal.speak()
dog.speak()
```

## `super()` bij overriding

Met `super()` roep je de methode van de parent class aan.

Waarom?

- Je wil bestaand gedrag behouden.
- Je wil extra gedrag toevoegen voor of na het parent-gedrag.
- Je vermijdt dubbele code.

```python
class Animal:
    def speak(self):
        print("Animal speaks")


class Dog(Animal):
    def speak(self):
        super().speak()
        print("Dog barks")


my_dog = Dog()
my_dog.speak()
```

## `super()` in constructors

Bij inheritance heeft de child class vaak extra velden. De gemeenschappelijke velden laat je door de parent constructor invullen.

```python
class Animal:
    def __init__(self, age):
        self.age = age


class Dog(Animal):
    def __init__(self, age, breed):
        super().__init__(age)
        self.breed = breed

    def display_info(self):
        print(f"Dog of age {self.age} and breed {self.breed}")


my_dog = Dog(age=3, breed="Labrador")
my_dog.display_info()
```

## Superclass properties gebruiken

Als de parent class een property heeft, kan de child class die gewoon gebruiken. Intern blijft de validatie of opslag bij de parent class.

```python
class Animal:
    def __init__(self, age):
        self.age = age

    @property
    def age(self):
        return self.__age

    @age.setter
    def age(self, value):
        if value < 0:
            raise ValueError("Age cannot be negative")
        self.__age = value


class Dog(Animal):
    def __init__(self, age, breed):
        super().__init__(age)
        self.breed = breed

    def get_age_in_human_years(self):
        return self.age * 7


my_dog = Dog(age=3, breed="Labrador")
print(my_dog.get_age_in_human_years())
```

## Method Resolution Order (MRO)

MRO is de volgorde waarin Python zoekt naar methodes bij inheritance.

Bij simpele inheritance zoekt Python eerst in de child class en daarna in de parent class. `super()` volgt die volgorde.

```python
class A:
    def show(self):
        print("A")


class B(A):
    def show(self):
        print("B before")
        super().show()
        print("B after")


b = B()
b.show()
print(B.mro())
```
