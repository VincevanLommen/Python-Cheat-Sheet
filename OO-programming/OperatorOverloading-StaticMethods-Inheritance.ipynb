## Operator overloading

Operator overloading betekent dat je bepaalt wat een operator doet voor objecten van je eigen klasse.

Waarom?

- Je objecten voelen natuurlijker aan in Python.
- `vector1 + vector2` is duidelijker dan `vector1.add(vector2)`.
- Je kan gedrag koppelen aan bekende Python-operatoren.

Hoe?

Je implementeert dunder methods zoals:

- `__add__` voor `+`
- `__sub__` voor `-`
- `__mul__` voor `*`
- `__str__` voor `print()` / `str()`

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        if not isinstance(other, Vector):
            raise TypeError("Can only add Vector to Vector")
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        if not isinstance(other, Vector):
            raise TypeError("Can only subtract Vector from Vector")
        return Vector(self.x - other.x, self.y - other.y)

    def __mul__(self, multiplier):
        if not isinstance(multiplier, (int, float)):
            raise TypeError("Multiplier must be a number")
        return Vector(self.x * multiplier, self.y * multiplier)

    def __str__(self):
        return f"Vector({self.x}, {self.y})"


v1 = Vector(1, 2)
v2 = Vector(3, 4)

print(v1 + v2)
print(v2 - v1)
print(v1 * 5)
```

## `__str__`

`__str__` bepaalt hoe een object eruitziet als je `print(object)` of `str(object)` gebruikt.

Zonder `__str__` krijg je meestal iets zoals `<__main__.Vector object at ...>`. Dat is technisch juist, maar niet nuttig voor mensen.

```python
class Money:
    def __init__(self, amount, currency):
        self.amount = amount
        self.currency = currency

    def __add__(self, other):
        if not isinstance(other, Money):
            raise TypeError("Can only add Money to Money")
        if self.currency != other.currency:
            raise ValueError("Cannot add different currencies")
        return Money(self.amount + other.amount, self.currency)

    def __str__(self):
        return f"{self.amount:.2f} {self.currency}"


wallet = Money(10, "EUR")
refund = Money(4.5, "EUR")
print(wallet + refund)
```

## Static methods

Een static method hoort logisch bij een klasse, maar gebruikt geen `self` en geen objectdata.

Waarom?

- Je groepeert helperlogica bij de juiste klasse.
- Je hoeft geen object te maken om de methode te gebruiken.
- De methode werkt enkel met de argumenten die je meegeeft.

Hoe?

Gebruik `@staticmethod` boven de methode.

```python
class FootballTeam:
    def __init__(self, name, points):
        self.name = name
        self.points = points

    @staticmethod
    def compare_teams(team1, team2):
        return team1.points > team2.points


team_a = FootballTeam("Team A", 15)
team_b = FootballTeam("Team B", 12)

print(FootballTeam.compare_teams(team_a, team_b))
```

## Static method als validator

Een vaak gebruikte toepassing is validatie. De validatieregel hoort bij de klasse, maar heeft geen bestaand object nodig.

```python
class Ticket:
    def __init__(self, ticket_id):
        self.ticket_id = ticket_id

    @property
    def ticket_id(self):
        return self.__ticket_id

    @ticket_id.setter
    def ticket_id(self, value):
        if not Ticket.is_valid_ticket_id(value):
            raise ValueError("Invalid ticket id")
        self.__ticket_id = value

    @staticmethod
    def is_valid_ticket_id(value):
        return (
            isinstance(value, str)
            and len(value) == 8
            and value[:3].isupper()
            and value[:3].isalpha()
            and value[3:].isdigit()
        )


ticket = Ticket("VIP12345")
print(ticket.ticket_id)
print(Ticket.is_valid_ticket_id("abc12345"))
```

## Inheritance

Inheritance betekent dat een child class eigenschappen en methodes erft van een parent class.

Waarom?

- Gemeenschappelijke code staat maar een keer in de parent class.
- Subclasses kunnen specialiseren wat anders is.
- Je modelleert een duidelijke `is-a` relatie: een `Circle` is een `Shape`.

Hoe?

Zet de parent class tussen haakjes: `class Circle(Shape):`. Gebruik `super()` om de constructor of methodes van de parent class aan te roepen.

```python
class Shape:
    def __init__(self, color):
        self.color = color

    def describe(self):
        return f"This is a {self.color} shape."


class Circle(Shape):
    def __init__(self, color, radius):
        super().__init__(color)
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

    def describe(self):
        return f"This is a {self.color} circle with radius {self.radius}."


class Square(Shape):
    def __init__(self, color, side_length):
        super().__init__(color)
        self.side_length = side_length

    def area(self):
        return self.side_length ** 2

    def describe(self):
        return f"This is a {self.color} square with side {self.side_length}."


shapes = [Circle("red", 3), Square("blue", 4)]

for shape in shapes:
    print(shape.describe())
    print(shape.area())
```
